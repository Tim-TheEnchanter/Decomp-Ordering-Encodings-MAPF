```python
map_lines  = input_map.readlines()
scen_lines = input_scen.readlines()
```

```python
scen_lines[:5]
```

```python
numbers_scen = []

for line in scen_lines[1:]:
    numbers_scen.append(re.findall(r'(\d+(?:\.\d+)?)', line)[-5:-1]) # saves the coordinates of agents and their goals in an array
```

```python
numbers_scen
```

the coordinates represent the staring coordinates SX and SY and the goal coordinates GX and GY.
The converted line has to follow the form of:
> init(object(robot,agent+1), value(at,(SX,SY))).

and

> init(object(shelf,agent+1), value(at,(GX,GY))).


```python
output_file = open("/home/tim/Documents/Uni/Bachelor Thesis/instances/maze-32-32-2.even-10scen.lp","a")

for agent in range(len(numbers_scen)):
    
    SX = int(numbers_scen[agent][0])+1
    SY = int(numbers_scen[agent][1])+1
    GX = int(numbers_scen[agent][2])+1
    GY = int(numbers_scen[agent][3])+1  
    converted_lines = "init(object(robot,%d), value(at,(%s,%s))).\ninit(object(destination,%d), value(at,(%s,%s))).\n\n" %(agent+1,SX,SY,agent+1,GX,GY) 
    output_file.write(converted_lines)
```

```python
output_file.close()
```
