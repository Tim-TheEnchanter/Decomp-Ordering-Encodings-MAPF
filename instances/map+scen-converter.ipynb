```python
import os

directory_maps  = os.fsencode("/home/tim/Documents/Uni/Bachelor Thesis/instances/maps/")
directory_scens = os.fsencode("/home/tim/Documents/Uni/Bachelor Thesis/instances/scenarios/")
directory_target = os.fsencode("/home/tim/Documents/Uni/Bachelor Thesis/instances/composed instances/")

for map_file in os.listdir(directory_maps):
    map_name = os.fsdecode(map_file)
    if map_name.endswith(".map"):
        print(map_name[:-4])
        
              
        for scen_file in os.listdir(directory_scens):
            scen_name = os.fsdecode(scen_file)
            if scen_name.startswith(map_name[:-4]):
                
                # converts .map to .lp
                subprocess.run(["map-to-lp","-i",directory_maps+map_file,"-o",directory_target+os.fsencode(scen_name[:-5]+".lp")])
         
                
                print("\t", scen_name)
                output_file = open(os.fsdecode(directory_target)+scen_name[:-5]+".lp", "a")
                scen        = open(os.fsdecode(directory_scens)+scen_name, "r")
                
                scen_lines = scen.readlines()
                
                numbers_scen = []
                
                for line in scen_lines[1:]:
                    numbers_scen.append(re.findall(r'(\d+(?:\.\d+)?)', line)[-5:-1]) # saves the coordinates of agents and their goals in an array
                
                for agent in range(len(numbers_scen)):
    
                    SX = int(numbers_scen[agent][0])+1
                    SY = int(numbers_scen[agent][1])+1
                    GX = int(numbers_scen[agent][2])+1
                    GY = int(numbers_scen[agent][3])+1  
                    converted_lines = "init(object(robot,%d), value(at,(%s,%s))).\ninit(object(destination,%d), value(at,(%s,%s))).\n\n" %(agent+1,SX,SY,agent+1,GX,GY) 
                    output_file.write(converted_lines)
                
                output_file.close()
```
